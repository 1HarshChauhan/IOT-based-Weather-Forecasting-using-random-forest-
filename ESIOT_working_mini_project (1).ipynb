```python
#include <DHT.h>
#include <Wire.h>
#include <LiquidCrystal_I2C.h>

// Define DHT22 sensor
#define DHTPIN 2          // Digital pin connected to the DHT22
#define DHTTYPE DHT22     // DHT 22 (AM2302)
DHT dht(DHTPIN, DHTTYPE);

// Initialize the LCD, address 0x27 for a 20x4 LCD (change if necessary)
LiquidCrystal_I2C lcd(0x27, 20, 4);

void setup() {
  // Initialize Serial communication
  Serial.begin(9600);
  while (!Serial) {
    ; // Wait for serial port to connect (needed for some boards)
  }

  // Initialize DHT sensor
  dht.begin();

  // Initialize LCD
  lcd.init();
  lcd.backlight();

  // Display welcome message across 4 lines
  lcd.setCursor(0, 0);
  lcd.print("Weather Predictor");
  lcd.setCursor(0, 1);
  lcd.print("Initializing...");
  delay(2000);
  lcd.clear();
}

void loop() {
  // Read temperature and humidity
  float humidity = dht.readHumidity();
  float temperature = dht.readTemperature(); // Celsius

  // Check if any reads failed
  if (isnan(humidity) || isnan(temperature)) {
    Serial.println("ERROR");
    lcd.setCursor(0, 0);
    lcd.print("Sensor Error       "); // Extra spaces to clear previous characters
    lcd.setCursor(0, 1);
    lcd.print("Check Connections ");
    return;
  }

  // Send data to Python in "temperature,humidity" format
  Serial.print(temperature);
  Serial.print(",");
  Serial.println(humidity);

  // Wait for Python's response
  if (Serial.available() > 0) {
    String prediction = Serial.readStringUntil('\n');
    prediction.trim(); // Remove any trailing newline or whitespace

    // Display prediction on LCD
    lcd.clear();
    lcd.setCursor(0, 0);
    lcd.print("Temp: ");
    lcd.print(temperature, 1);
    lcd.print(" C");

    lcd.setCursor(0, 1);
    lcd.print("Humidity: ");
    lcd.print(humidity, 1);
    lcd.print(" %");

    lcd.setCursor(0, 2);
    lcd.print("Forecast:");
    
    lcd.setCursor(0, 3);
    lcd.print(prediction);
  }

  // Wait before next reading
  delay(5000); // 5 seconds
}
```

```python
import serial
import time
import requests
import pandas as pd
import numpy as np

# Define weather conditions
weather_conditions = np.array([
    'Partly Cloudy', 'Partly cloudy', 'Sunny', 'Light drizzle',
    'Clear', 'Overcast', 'Moderate or heavy rain with thunder',
    'Patchy light rain in area with thunder', 'Fog',
    'Patchy light rain with thunder', 'Patchy rain nearby',
    'Light rain', 'Mist', 'Cloudy', 'Thundery outbreaks in nearby',
    'Moderate rain', 'Light rain shower', 'Moderate rain at times',
    'Moderate or heavy rain shower', 'Heavy rain at times',
    'Patchy light rain', 'Patchy light drizzle',
    'Thundery outbreaks possible', 'Patchy rain possible',
    'Moderate or heavy rain in area with thunder', 'Heavy rain',
    'Torrential rain shower', 'Freezing fog',
    'Moderate or heavy snow showers'
], dtype=object)

# Constants
API_KEY = '2a9ed43bc339775fa67f4e89ad61b593'  # Replace with your OpenWeather API key
CITY = 'Kothrud'            # Replace with your desired city
BASE_URL = 'https://api.openweathermap.org/data/2.5/weather'

SERIAL_PORT = 'COM4'  # Replace with your Arduino's serial port (e.g., 'COM4' on Windows or '/dev/ttyACM0' on Linux/Mac)
BAUD_RATE = 9600

def get_additional_weather_data():
    url = f"{BASE_URL}?q={CITY}&appid={API_KEY}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Error: Unable to fetch data for {CITY}. Status code: {response.status_code}")
        return None

def extract_additional_info(data):
    if data:
        wind = data.get('wind', {})
        clouds = data.get('clouds', {})
        visibility = data.get('visibility', 10000)
        main = data.get('main', {})

        additional_info = {
            'wind_kph': wind.get('speed', 0) * 3.6,
            'wind_degree': wind.get('deg', 0),
            'pressure_mb': main.get('pressure', 1013),
            'cloud': clouds.get('all', 0),
            'visibility_km': visibility / 1000,
        }

        return additional_info
    return None

def weather_prediction(sensor_info):
    api_data = get_additional_weather_data()
    additional_info = extract_additional_info(api_data)

    if additional_info is None:
        return "API Error"

    combined_info = {
        'temperature_celsius': sensor_info['temperature_celsius'],
        'wind_kph': additional_info['wind_kph'],
        'wind_degree': additional_info['wind_degree'],
        'pressure_mb': additional_info['pressure_mb'],
        'humidity': sensor_info['humidity'],
        'cloud': additional_info['cloud'],
        'visibility_km': additional_info['visibility_km']
    }

    input_values = pd.DataFrame([[
        combined_info['temperature_celsius'],
        combined_info['wind_kph'],
        combined_info['wind_degree'],
        combined_info['pressure_mb'],
        combined_info['humidity'],
        combined_info['cloud'],
        combined_info['visibility_km']
    ]], columns=['temperature_celsius', 'wind_kph', 'wind_degree',
                  'pressure_mb', 'humidity', 'cloud', 'visibility_km'])

    # Make prediction (Assuming 'rf' is your trained model)
    prediction = rf.predict(input_values)[0]
    predicted_condition = weather_conditions[prediction]
    print(f"Forecasted Weather: {predicted_condition}")
    return predicted_condition

def main():
    try:
        ser = serial.Serial(SERIAL_PORT, BAUD_RATE, timeout=1)
        time.sleep(2)
        print(f"Connected to {SERIAL_PORT} at {BAUD_RATE} baud.")
    except serial.SerialException:
        print(f"Could not open serial port {SERIAL_PORT}.")
        return

    try:
        i=0
        while i<5:
            i=i+1
            if ser.in_waiting > 0:
                try:
                    line = ser.readline().decode('utf-8').strip()
                except UnicodeDecodeError:
                    print("Received invalid data. Skipping...")
                    continue

                if line == "ERROR":
                    print("Received ERROR from sensor.")
                    continue

                try:
                    temp_str, hum_str = line.split(',')
                    temperature = float(temp_str)
                    humidity = float(hum_str)
                    print(f"Received -> Temperature: {temperature}Â°C, Humidity: {humidity}%")

                    sensor_info = {
                        'temperature_celsius': temperature,
                        'humidity': humidity
                    }

                    prediction = weather_prediction(sensor_info)

                    # Send prediction back to Arduino
                    ser.write((prediction + "\n").encode())

                except ValueError:
                    print(f"Invalid data received: {line}")
            time.sleep(1)
    except KeyboardInterrupt:
        print("\nExiting program.")
    finally:
        ser.close()

if __name__ == "__main__":
    main()
```
